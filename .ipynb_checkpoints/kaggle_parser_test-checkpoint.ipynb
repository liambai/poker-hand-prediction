```python
import sqlite3

conn = sqlite3.connect('poker.db')
c = conn.cursor()
```

```python
lines = open("kaggle-data/kaggle_file1.txt" ,"r").readlines()
games = [[]]
for line in lines:
    games[-1].append(line)
    if line == '\n':
        games.append([])
```

```python
for game in games:
    game_id = game[1].split(' ')[2]
    stakes = game[1].split(' ')[3]
    big_blind = stakes.split("/")[1]
    for line in game:
        words = line.split(" ")
        
        #get info about the 5 cards on the board
        if words[0] == "Board:":
            cards = [None]*5
            cards_info = words[1:]
            for i in range(len(cards_info)):
                cards[i] = cards_info[i].strip("[]")
    
    c.execute('''INSERT INTO Games VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)''', \
              (game_id, big_blind, cards[0], cards[1], cards[2], \
               cards[3], cards[4], None, None, None, None))
    conn.commit()
```

```python
for game in games:
    game_id = game[1].split(' ')[2]
    players = set()
    round_k = 0
    actions = [] #each action: (player_id, round_k, pos_in_round, action_id)
    pos_in_round = 0
    for line in game:
        words = line.split(" ")
        if words[0] == "***":
            round_k += 1
            pos_in_round = 0
        if words[0] == "Player":
            player_id = words[1]
            if player_id not in players:
                players.add(player_id)
            action_word = words[2].strip("\n ")
            if action_word == 'folds':
                pos_in_round += 1
                actions.append((player_id, round_k, pos_in_round, 'f'))
                
    
```
